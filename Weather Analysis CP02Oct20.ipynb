```python
#needed to make web requests
import requests
import pandas as pd
import json
import numpy as np

from datetime import datetime
```

```python
file = "Resources/weather_temp_history2019_2020.csv"
city_temp_data = pd.read_csv(file)
city_temp_data.info()
```

```python
city_temp_data
```

```python
cities = city_temp_data['dt_iso'].astype(str)
dates = []
for city in cities:
    dates.append(city[:-19])
    
#print(dates)
city_temp_data['dt_iso']= dates

city_temp_data.head()
```

```python
city_data_groupby_df = city_temp_data.groupby(['dt_iso','city_name']).mean()
```

```python
city_data_groupby_df.head()
```

```python
final_city_temp_data_groupby = city_data_groupby_df.loc['2020-03-18':'2020-08-15']
final_city_temp_data_groupby.head()
```

```python
city_temp_data_df = pd.DataFrame(final_city_temp_data_groupby)
city_temp_data_df.to_csv('Resources/Chicago_LA_output_city_temp__df.csv')
```
